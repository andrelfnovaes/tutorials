```python
import pandas as pd
import numpy as np

n_rows = n_cols = 10

data = np.arange(1, (n_rows*n_cols)+1).reshape(n_rows,n_cols).tolist()
index = ['row '+str(i) for i in range(n_rows)]
columns = [chr(97+i) for i in range(n_rows)]

df = pd.DataFrame(
    data = data,
    index = index,
    columns = columns,
)
display(df)
```
